```python
from dostoevsky.tokenization import RegexTokenizer
from dostoevsky.models import FastTextSocialNetworkModel
# https://github.com/bureaucratic-labs/dostoevsky

tokenizer = RegexTokenizer()
model = FastTextSocialNetworkModel(tokenizer=tokenizer)

files = [
    'mumu.txt', # Негативно
    'prestuplenie-i-nakazanie.txt',  # Негативно
    'slovo-patsana.txt',  # Нейтрально
    'happy-new-year.txt',  # Формальное поздравление
    'kitten.txt' # Позитивно
]


for filee in files:
  f = open(filee,'r').read().replace("\n", " ") # Читаем файл и переводим все в одну большую строку
  print(f)
  result = model.predict([f], k=2)
  print(result)

```
